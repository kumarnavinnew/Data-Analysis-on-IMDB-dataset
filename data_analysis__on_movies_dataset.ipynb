# 20. Classify Movies Based on Ratings [Excellent, Good, and Average]

```python
def rat(rating):

    if rating>=7:
      return "Excellent"
    elif rating >=6:
      return "Good"
    else:
      return "Average"
df["rating_category"]=df["Rating"].apply(rat)
df[["Title","rating_category"]]
```

# 23. How Many Films of Each Genre Were Made?

```python

df['Genre'] = df['Genre'].str.strip()
df["Genre"]
df['Genre'] = df['Genre'].str.split(',')

# Explode the dfFrame to have one genre per row
df_exploded = df.explode('Genre')
df_exploded
# Count the number of films for each genre
genre_counts = df_exploded['Genre'].value_counts()

print(genre_counts)
```
